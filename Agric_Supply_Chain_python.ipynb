```python
## import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## load csv file
agric_supply_chain = pd.read_csv("agric_supply_chain.csv")
agric_supply_chain
```

```python
## make a copy
df = agric_supply_chain.copy()
```

```python
## check duplicates
df.duplicated().sum()
```

```python
## check the data types
df.dtypes
```

```python
## data summary
df.describe()
```

```python
## convert harvest date to date time
df["harvest_date"] = pd.to_datetime(df["harvest_date"])
```

```python
## calculate the price margin
df["price_margin"] = df["market_price_per_kg"] - df["farm_gate_price_per_kg"]
df["price_margin"]
```

```python
## calculate the net profit
df["net_profit"] = df["gross_revenue_ngn"] - df["transport_cost_ngn"]
df["net_profit"]
```

```python
## calculate the average post-harvest loss % by crop and visualize it with bar chart
Average_post_harvest_loss_by_crop = df.groupby("crop")["post_harvest_loss_pct"].mean().sort_values()
Average_post_harvest_loss_by_crop.plot(kind = "bar", figsize=(10,6))
plt.title("Average_post_harvest_loss_by_crop.plot")
plt.ylabel("Average Loss%")
plt.xlabel("Crop")
plt.show()
```

```python
## visualize and the calculate total gross revenue by state
Total_gross_revenue_by_state = df.groupby("state")["gross_revenue_ngn"].sum().sort_values()
Total_gross_revenue_by_state.plot(kind = "bar", figsize=(8,6))
plt.title("Total gross revenue by state")
plt.ylabel("Total Gross Reveneue")
plt.xlabel("State")
plt.show()
```

```python
## Line chart: Average market price per kg by month and year
df["year_month"] = df["harvest_date"].dt.to_period("M")
Avg_mkt_price_by_month_and_year = (
    df.groupby("year_month")["market_price_per_kg"].mean())
Avg_mkt_price_by_month_and_year.plot(kind="line", figsize=(12,6), marker="o")
plt.title("Average Market Price per KG by Month")
plt.ylabel("Average Market Price")
plt.xlabel("Month")
plt.xticks(rotation = 45)
plt.show()
```

```python
## Avg post-harvest loss — cooperative members vs non-members
cooperative_loss = df.groupby("cooperative_member")["post_harvest_loss_pct"].mean()
cooperative_loss.plot(kind="bar", figsize=(10,6))
plt.title("Average Post-Harvest Loss by Cooperative Membership")
plt.ylabel("Average Post-Harvest Loss")
plt.xlabel("Cooperative Member")
```

```python
## Transport mode vs average days to market
avg_transport_mode_days_to_mkt = df.groupby("transport_mode")["days_to_market"].sum().sort_values()
avg_transport_mode_days_to_mkt.plot(kind = "bar", figsize=(10,6))
plt.title("Average Days to Market by Transport Mode")
plt.ylabel("Days")
plt.xlabel("Transport Mode")
plt.show()
```

```python
## Price margin by crop
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="crop", y="price_margin")
plt.title("Price Margin by Crop")
plt.xticks(rotation=45)
plt.show()
```

```python
## Heatmap: Avg net profit by state and crop
avg_profit_heatmap = pd.pivot_table(df, values="net_profit", index="state", columns="crop", aggfunc="mean")
plt.figure(figsize=(10,6))
sns.heatmap(avg_profit_heatmap, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Net Profit by State and Crop")
plt.show()
```

```python
## Bar chart: Revenue by storage type
revenue_by_storage_type = df.groupby("storage_type")["gross_revenue_ngn"].sum().sort_values()
revenue_by_storage_type.plot(kind="bar", figsize=(10,6))
plt.title("Revenue By Storage Type")
plt.ylabel("Revenue")
plt.xlabel("Storage Type")
plt.show()
```

### Summary: 5 supply chain improvement recommendations
Key Recommendations

2. Encourage cooperative membership to enhance access to resources and market opportunities.
3. Optimize transportation systems to reduce delays and improve market access.
4. Promote cultivation of high-margin crops to increase farmer profitability.
5. Strengthen market information systems to support better pricing and sales decisions.
